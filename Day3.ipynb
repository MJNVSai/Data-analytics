### LCR Circuit Graph

```python
# to draw the graph of the LCR circuit
import matplotlib.pyplot as plt

f = [s for s in range(4,25,1)]
# here values of 'f' are frequencies

r200 = [2.02,2.35,2.62,3.18,4.36,5.20,6.71,8.31,10.42,12.26,13.75,12.13,10.96,8.98,7.51,6.02,5.18,4.37,3.98,3.44,3.14]
# r200 is the resistance of 200 ohms  for current

r500 = [1.74,2.15,2.75,3.40,3.78,4.65,5.85,6.61,7.36,7.60,7.85,7.32,6.68,6.08,5.56,5.05,4.32,3.92,3.57,3.19,2.85]
# r500 is the resistance of 500 ohms for current

plt.plot(f,r200,"ro")
plt.plot(f,r200)

plt.plot(f,r500,'bs')
plt.plot(f,r500)

plt.xlabel('frequency in HZ')
plt.ylabel('current in mA')

plt.title('LCR Ciruit')
plt.show()
```

```python
vals = [num for num in range(10,50,5)]

plt.hist(std1,vals, histtype='bar',orientation='vertical',color='b')

plt.show()
```

```python
vals = [num for num in range(10,50,5)]

plt.hist(std1,vals, histtype='bar',orientation='vertical',color='b')
plt.hist(std2,bins=10,color='g',orientation='horizontal')
plt.show()
```

```python
from matplotlib import style

style.use('ggplot')

plt.figure(figsize=(8,5))
plt.hist([std1,std2],vals, rwidth=0.7,histtype='bar',orientation='vertical',color=['m','y'])
plt.title('Multi-Histograms')
plt.show()
```

### Pie chart plot (pie-plot)

```python
x = [56,25,30,24] # making the regions

plt.pie(x)
plt.show()
```

```python
# adding labels

fb = ['average','good','above average','exellent']
plt.pie(x,labels=fb) # assingnig labels of fb to x
plt.show()
```

```python
# addings colors to regions of pie
plt.pie(x,labels=fb, colors=['green','cyan','blue','pink'], autopct="%.f%%")

plt.title(' Sort by Responses')
plt.show()
```

```python
# addings shadow to regions of pie
plt.pie(x,labels=fb, colors=['green','cyan','blue','pink'], autopct="%.f%%", shadow = True, startangle=180, explode = [0,1,0,0])
# explode means exluding

plt.title(' Sort by Responses')
plt.show()
```
