### Merging 2 Data-Frames

```python
# merging 2 data frames

c = pd.DataFrame({'Int_rate':[2,1,2,3], 'Int_gdp':[50,45,45,67]}, index = [2041,2002,2003,2204])
print("result_1: \n\n",c)
```

```python
u = pd.DataFrame({'low_tier_hpi':[80,90,70,60], 'Unemployement':[1,3,5,6]}, index = [2001,2003,2003,2005])
print("result_2: \n\n",u)
```

```python
# joining the data-frames
res = c.join(u)
print("Joining the data- frames: \n\n",res)
```

```python
# merging the data-frames having similar column values

m = pd.DataFrame({'a':[4,5,6,7], 'b':[30,50,60,70], 'c':['a','b','c','d']}, index = ['first','second','third','fourth'])
print("Data-Frame: \n\n",m)
m
```

```python
n = pd.DataFrame({'a':[4,9,6,7], 'b':[30,80,60,70], 'c':['a','b','c','d']}, index = ['first','second','third','fourth'])
new = m.merge(n)
print("merged frame: \n\n",new)
m.merge(n)
```

```python
new.columns = ['f','s','th'] # renameing the column values
new.rows = [5,6,7] # updating the row values
print("modifyed frame: \n\n",new)
new
```

```python
new.columns = ['f','s','th'] # renameing the column values
new.index = [5,6,7] # updating the row values
print("modifyed frame: \n\n",new)
new
```

```python
# merging the dataframes based on particular column value

h = m.merge(n, on = 'c')
print("particular coulmn: \n\n",h)

m.merge(n, on = 'c')
```
